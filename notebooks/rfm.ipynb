```python
import pandas as pd
```

```python
clv_df = pd.read_csv("../data/processed/rfm_with_CLVclusters.csv")
```

```python
customers_df = clv_df.rename(
    columns={
        "Customer ID": "customer_id",
        "Frequency": "frequency",
        "Monetary": "monetary",
        "Lifespan": "lifespan",
        "Cluster": "clv_tier",
        "First Purchase Date": "first_purchase_date",
        "Last Purchase Date": "last_purchase_date"
    }
)
```

```python
customers_df["segment"] = None
customers_df["churn_probability"] = None
customers_df["last_updated"] = pd.Timestamp.now()
```

```python
customers_df = customers_df[
    [
        "customer_id",
        "frequency",
        "monetary",
        "lifespan",
        "segment",
        "clv_tier",
        "churn_probability",
        "first_purchase_date",
        "last_purchase_date",
        "last_updated"
    ]
]
```

```python
customers_df.to_csv(
    "../data/processed/customer_intelligence.csv",
    index=False
)

print(customers_df.head())
```
